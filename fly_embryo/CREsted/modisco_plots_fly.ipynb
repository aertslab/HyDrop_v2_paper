```python
import os

def match_h5_files_to_classes(contribution_dir: str, classes: list[str]) -> dict[str, str | None]:
    """
    Match .h5 files in a given directory with a list of class names and returns a dictionary mapping.

    Parameters
    ----------
    contribution_dir : str
        Directory containing .h5 files.
    classes : list[str]
        List of class names to match against file names.

    Returns
    -------
    dict[str, str | None]
        A dictionary where keys are class names and values are paths to the corresponding .h5 files if matched, None otherwise.
    """
    # Get all .h5 files in the directory
    h5_files = [file for file in os.listdir(contribution_dir) if file.endswith(".h5")]
    print(f"Found .h5 files: {h5_files}")  # Debugging: List all files found

    # Initialize dictionary with class names as keys and None as default value
    matched_files = {class_name: None for class_name in classes}

    # Loop through the files and match by stripping "_modisco_result.h5"
    for file in h5_files:
        base_name = file.replace("_modisco_result.h5", "")
        print(f"Processing file: {file}, extracted base name: {base_name}")  # Debugging: Show base name

        for class_name in classes:
            if base_name == class_name:
                matched_files[class_name] = os.path.join(contribution_dir, file)
                print(f"Matched {class_name} to {file}")  # Debugging: Show matches
                break
            else:
                print(f"No match for {class_name} with {base_name}")  # Debugging: Show non-matches

    return matched_files
```

```python
# First we obtain the resulting modisco files per class
matched_files = crested.tl.modisco.match_h5_files_to_classes(
    contribution_dir=contribution_dir, classes=celltype_list
)
```

```python
# Then we cluster matching patterns, and define a pattern matrix [#classes, #patterns] describing their importance
all_patterns = crested.tl.modisco.process_patterns(
    matched_files,
    sim_threshold=3,  # lowered to 3 (lower than in tutorial) to avoid seperating of patterns in heatmap
    trim_ic_threshold=0.05,  # Information content (IC) threshold on which to trim patterns
    discard_ic_threshold=0.2,  # IC threshold used for discarding single instance patterns
    verbose=True,  # Useful for doing sanity checks on matching patterns
)
pattern_matrix = crested.tl.modisco.create_pattern_matrix(
    classes=celltype_list,
    all_patterns=all_patterns,
    normalize=False,
    pattern_parameter="seqlet_count_log",
)
pattern_matrix.shape
```

```python
import os
import glob
import pandas as pd

# Path to the motif database
motif_db_path = "/user/leuven/350/vsc35050/hdickm/resources/motifs-v10nr_clust-nr.flybase-m0.001-o0.0.tbl"

# Columns to extract from the motif database
columns_to_extract = [
    "#motif_id",	"motif_name",
    "motif_description",
    "source_name",
    "source_version",
    "gene_name",
    "motif_similarity_qvalue",
    "similar_motif_id",
    "similar_motif_description",
    "orthologous_identity",
    "orthologous_gene_name",
    "orthologous_species",
    "description"
]

# Path to the directory containing motif files
input_dir = "/lustre1/project/stg_00002/lcb/hydrop_v2_paper/fly/CREsted/hydrop_Eren/modisco_results_3k_wf_output"

# Load the motif database
motif_db = pd.read_csv(motif_db_path, sep="\t")

# Ensure the required columns are in the motif database
missing_columns = [col for col in columns_to_extract + ["motif_name"] if col not in motif_db.columns]
if missing_columns:
    raise ValueError(f"The following required columns are missing in the motif database: {', '.join(missing_columns)}")

# Filter to retain only necessary columns
motif_db = motif_db[["motif_name"] + columns_to_extract]

# Iterate over all `_report` directories
report_dirs = glob.glob(os.path.join(input_dir, "*_report"))
for report_dir in report_dirs:
    # Path to the motifs.tsv file
    motifs_tsv_path = os.path.join(report_dir, "motifs.tsv")

    # Check if the motifs.tsv file exists
    if not os.path.exists(motifs_tsv_path):
        print(f"File not found: {motifs_tsv_path}. Skipping.")
        continue

    # Read the motifs.tsv file
    motifs_df = pd.read_csv(motifs_tsv_path, sep="\t")

    # Ensure the required columns "match0", "match1", "match2" exist in the motifs.tsv file
    missing_match_columns = [col for col in ["match0", "match1", "match2"] if col not in motifs_df.columns]
    if missing_match_columns:
        print(f"Columns {', '.join(missing_match_columns)} not found in {motifs_tsv_path}. Skipping.")
        continue

    # Merge motifs.tsv with the motif database for each match
    merged_dfs = []
    for match_col in ["match0", "match1", "match2"]:
        merged_df = pd.merge(motifs_df, motif_db, how="left", left_on=match_col, right_on="Motif_name")
        merged_dfs.append(merged_df)

    # Concatenate annotations across match0, match1, match2 for each column
    concatenated_annotations = {}
    for col in columns_to_extract:
        concatenated_annotations[col] = merged_dfs[0][col].astype(str).fillna("") + ";" + \
                                    merged_dfs[1][col].astype(str).fillna("") + ";" + \
                                    merged_dfs[2][col].astype(str).fillna("")
    # Remove empty annotations, 'nan', and deduplicate
        concatenated_annotations[col] = concatenated_annotations[col].apply(
          lambda x: ";".join(sorted(set(filter(lambda y: y and y != "nan", x.split(";")))))
      )

# Add concatenated annotations to the original DataFrame
    for col, concatenated_col in concatenated_annotations.items():
      motifs_df[col] = concatenated_col

    # Save the annotated file
    output_path = os.path.join(report_dir, "motifs_annot.tsv")
    motifs_df.to_csv(output_path, sep="\t", index=False)
    print(f"Annotated motifs saved to: {output_path}")

print("Annotation process completed.")
```
